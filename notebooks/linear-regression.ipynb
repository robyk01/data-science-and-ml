```python
import numpy as np
import matplotlib.pyplot as plt

# Let's say X is 'Hours Studied' and y is 'Exam Score'
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([15, 25, 35, 42, 50, 62])

plt.scatter(X, y) # This shows our raw data points
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# We create an instance of the model. 
model = LinearRegression()
```

```python
# This is the learning step.
model.fit(X, y)

m = model.coef_      # The Slope
b = model.intercept_ # The Intercept

print(f"The equation is: y = {m[0]:.2f}x + {b:.2f}")
```

```python
new_student = np.array([[4.5]])
prediction = model.predict(new_student)

print(f"Predicted score for 4.5 hours: {prediction[0]:.2f}")
```

```python
score = model.score(X, y)
print(f"R-squared: {score:.4f}")
```

```python
# Plot the actual data points
plt.scatter(X, y, color='blue', label='Actual Data')

# Use the model to predict y-values for our X range to draw the line
y_pred = model.predict(X)

# Plot the prediction line
plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression Line')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression: Study Hours vs Score')
plt.legend()

plt.show()
```
